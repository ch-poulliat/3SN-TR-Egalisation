## Filtrage RIF et convolution
Sur la figure, $x[n]$ est la séquence d'entrée d'un filtre linéaire invariant discret de réponse impulsionnelle $h[n] .$ Chaque séquence est de longueur finie. 

```matlab
h=[0 0 2 1 -0.5 0 0 0];
x=[0 0 2 2 2 2 0 0];
figure
stem((-2:5),x);
grid on;
xlabel('n');
title('x[n]');
figure
stem((-2:5),h,'r');
grid on;
xlabel('n');
title('h[n]');
```

1.1 Déterminer graphiquement la réponse $y[n]$ du filtre.

```matlab
y=conv(h,x);

stem((-4:10),y);
xlabel('n');
title('y[n]');
```

1.2 Ecrire son équation aux différences.

Le filtrage étant un filtrage de type RIF, l'équation aux récurrences s'identifie grâce au produit de convoltion entre la réponse impulsionnelle et une entrée $x[n]$. Ainsi
$$y[n]= h[n]*x[n] =\sum_{k \in \mathbb{Z}} h[k] x[n-k]=h[0] x[n] + h[1] x[n-1] + h[2] x[n-2]$$ et donc

$$y[n] =2 x[n] + x[n-1] -0.5 x[n-2].$$

*NB : le transfert associé est donné par $$\bar{h}(z)=\frac{\bar{y}(z)}{\bar{x}(z)}=2+z^{^-1}-0.5z^{^-2}$$* 

1.3 Quelle est la longueur de la convolution de 2 séquences finies de longueur $\mathrm{M}$ et $\mathrm{N}?$

De manière triviale, on trouve $M+N-1.$

2.1 Soit la fraction rationnelle $\overline{\mathrm{h}}_{1}(\mathrm{z})=\frac{1}{1-\mathrm{a} \cdot \mathrm{z}^{-1}}$
Cette fonction de la variable complexe $z$ est la transformée en $\mathcal{Z}$ de 2 séquences, $\mathrm{h}_{11}[\mathrm{n}]$ causale et $\mathrm{h}_{12}[\mathrm{n}]$ non causale.
Déterminer $\mathrm{h}_{11}[\mathrm{n}]$ et $\mathrm{h}_{12}[\mathrm{n}]$ et préciser les domaines de convergence des transformées en $\mathcal{Z}$ associées.

En utilisant le développement en série de 
$\frac{1}{1-x}= \sum_n x^n, \; \forall |x|<1,$

on trouve deux cas :

$(a)$ : $|az^{^-1}| <1 \leftrightarrow \textrm{ROC}_1 : |a| < |z| $ 

dans ce cas, on a alors

$$\bar{h}_1{(z)}=\sum_{k=0}^{+\infty}{a^n z^{^-n}}$$

Par transformée inverse, on trouve 

$$h_1[n]=a^n,\forall n \geq 0, 0 \;\mathrm{ sinon}.$$

$(b)$ : $|az^{^-1}| >1 \leftrightarrow \textrm{ROC}_2 : |a| > |z| $ 

dans ce cas, on a alors

$$\bar{h}_2{z}=-\frac{a^{^-1}z}{1-a^{^-1}z}=-\sum_{k=1} a^{^-k}z^k$$

d'où $$h_2[n]=-a^n,\forall n < 0, 0 \; \mathrm{ sinon}.$$

Parmi cette infinité de solutions, deux seulement ont une transformée en $\mathcal{Z}$ :
pour $\mathrm{K}=0,$ la séquence causale $\mathrm{h}_{\mathrm{c}}[\mathrm{n}]=\mathrm{a}^{\mathrm{n}} . \mathrm{U}[\mathrm{n}]$ a une transformée qui converge pour $|\mathrm{z}|>|\mathrm{a}|$
pour $\mathrm{K}=-1,$ la séquence anticausale $\mathrm{h}_{\mathrm{a}}[\mathrm{n}]=-\mathrm{a}^{\mathrm{n}} . \mathrm{U}[-1-\mathrm{n}]$ a une transformée qui converge pour $|\mathrm{z}|<|\mathrm{a}|$

2.6 On pose $z=e^{+j 2 \pi \nu}$ et on appelle $M$ l'affixe de $z$ dans le plan complexe d'origine $O$. Ecrire le module de $\stackrel{\circ}{h}_{1}(v)$ en fonction de $\mathrm{OM}$ et $\mathrm{AM},$ A étant un point du plan que l'on précisera. 

Le module de $\stackrel{\circ}{h}_{1}(v)$ est donné par

$$|\stackrel{\circ}{h}_{1}(v)|=\frac{1}{\sqrt{1+|a|^2-2\Re{(a e^{-j2\pi\nu})}}}.$$

L'interprétation de $$\bar{h}_1(z)=\frac{z - 0}{z - a}$$ a l'aide des vecteurs d'affixes $z$ et $z-a$,
permet d'interpréter 
$$|\stackrel{\circ}{h}_{1}(v)|=\frac{OM}{MA}$$

où $M$ est le point d'affixe $e^{-j2\pi\nu}$, $O$ le point à l'origine et $A$ le point d'affixe $z=a.$

```matlab
a=0.5*exp(+1i*2*pi*0.33);
m=exp(+1i*2*pi*0.15);
zplane(1,[1,-a]);
hold on
plot(cos(2*pi*0.15),sin(2*pi*0.15),'rs')
%vectors
x=(0:cos(2*pi*0.15)/100:cos(2*pi*0.15));
y=(0:sin(2*pi*0.15)/100:sin(2*pi*0.15));
plot(x,y);

coef_dir=(imag(m)-imag(a))/(real(m)-real(a));
b=imag(a)-coef_dir*real(a);
x=(real(a):(real(m)-real(a))/100:real(m));
y=coef_dir*x+b;
plot(x,y,'r--');
%text
text(-0.1,-0.1,'O')
text(1.15*real(a),1.15*imag(a),'A')
text(1.1*cos(2*pi*0.15),1.1*sin(2*pi*0.15),'M')
```
